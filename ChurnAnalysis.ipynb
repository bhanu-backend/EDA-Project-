## ðŸ“‰ Objectives of Churn Analysis

- **Churn Distribution**
  - Bar graph (count plot) using `Churn`
  - Helps identify the proportion of churned vs retained customers

- **Churn Distribution by Gender**
  - Count plot with `Churn` and `gender`
  - Analyzes churn behavior across different genders

- **Numerical Feature Analysis**
  - Box and whisker plots using:
    - `Churn` vs `tenure`
    - `Churn` vs `MonthlyCharges`
    - `Churn` vs `TotalCharges`
  - Compares distributions and detects outliers

- **Correlation Analysis**
  - Heat map using `tenure`, `MonthlyCharges`, `TotalCharges`, `SeniorCitizen`, and `Churn`
  - Identifies relationships between numerical features and churn

- **Behavioral Pattern Analysis**
  - Scatter plots using:
    - `tenure` vs `MonthlyCharges` (hue: `Churn`)
    - `tenure` vs `TotalCharges` (hue: `Churn`)
  - Highlights churn-prone customer segments


```python
#Churn Distribution

plt.figure(figsize = (5,4))
ax = sbn.countplot(x = df["Churn"] , data = df )
plt.bar_label(ax.containers[0])
plt.title("Churn Distribution")
plt.ylabel("No. Of Customers")
plt.xlabel("Churn")
plt.plot()
```

```python
#Churn Distribution by gender 

plt.figure(figsize = (5,4))
ax = sbn.countplot(x = df["Churn"] , data = df ,hue = "gender" )
plt.title("Churn Distribution")
plt.ylabel("No. Of Customers")
plt.xlabel("Churn")
plt.plot()
```

As clearly seen, both **male and female customers churn in almost equal proportions**, indicating that **gender does not have a significant impact on customer churn** in this dataset.

```python

X = ["tenure" , "MonthlyCharges" , "TotalCharges" ]
for i in X :
     plt.figure(figsize = (6,2))
     plt.title(f"Churn Frequency Distribution Using Histrogram On {i}")
     sbn.histplot(x = i , data = df  , hue = "Churn" )
     plt.ylabel("Customer Count")
     plt.show()
```

## ðŸ“Š Tenure vs Churn â€” Summary

- Customers with **low tenure (0â€“12 months)** have a **higher likelihood of churn**.
- **Churn decreases as tenure increases**, indicating improved retention over time.
- **Long-tenure customers (50+ months)** are largely **loyal and stable**.
- Tenure is a **strong inverse indicator of churn**.

### âœ… Key Takeaway
Focusing retention efforts on **new customers during their early lifecycle** can significantly reduce overall churn.

```python
plt.figure(figsize = (5,5))
sbn.heatmap(
      corr_data,
      annot = True,
      fmt = "0.1f",
      cmap = "coolwarm")
plt.title("Correlation Using Heat Map")
plt.show()
```

###  Strong Correlation Insights & Actions

* **Tenure vs TotalCharges ( +0.80 )** shows a **strong positive correlation**, meaning customers who stay longer naturally generate higher total revenue.
* **Tenure vs Churn ( âˆ’0.40 )** represents the **strongest negative relationship**, indicating that churn decreases significantly as customer tenure increases.

### Course of Action

* Focus on **retaining customers during the early tenure period**, as this has the highest impact on reducing churn.
* Strengthen **long-term engagement programs** to increase tenure, which directly boosts revenue and lowers churn risk.
